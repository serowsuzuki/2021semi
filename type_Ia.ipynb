```python
import pandas as pd
import pandas_profiling
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("input/type_Ia.csv")
data01 = pd.read_csv("input/type_Ia_lt_001.csv")

print(data.count())
print(data01.count())
```

```python
data.dropna(subset=['z','dL (Mpc)'], inplace=True)
data01.dropna(subset=['z','dL (Mpc)'], inplace=True)

print(data.count())
print(data01.count())
```

```python
data01a = data01["z"].str.split(",", expand=True)
data01["z"] = data01a[0].astype(float)
data01.profile_report()
```

```python
x = data01["z"]
y = data01["dL (Mpc)"]
plt.scatter(x, y, s=3)
plt.title("sctter plot of z vs dL (Mpc)")
plt.rcParams["figure.figsize"] = [27,18]
plt.xlabel("z")
plt.ylabel("dL (Mpc)")
plt.grid()
plt.show()
```
